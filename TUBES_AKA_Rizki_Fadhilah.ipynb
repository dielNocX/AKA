#ALGORITMA ITERATIF DAN REKURSIF

###Iteratif

```python

def cek_matriks_diagonal_iteratif(matriks):
    n = len(matriks)
    for i in range(n):
        for j in range(n):
            if i != j and matriks[i][j] != 0:
                return False
    return True
```

###Rekursif

```python
def cek_matriks_diagonal_rekursif(matriks, i=0, j=0):
    if i == len(matriks):
        return True

    if j == len(matriks):
        return cek_matriks_diagonal_rekursif(matriks, i + 1, 0)

    if i != j and matriks[i][j] != 0:
        return False

    return cek_matriks_diagonal_rekursif(matriks, i, j + 1)
```

###WORST CASE

```python
import random
import time
import matplotlib.pyplot as plt

import sys


sys.setrecursionlimit(2000000000)

def cek_matriks_diagonal_iteratif(matriks):
    n = len(matriks)
    for i in range(n):
        for j in range(n):
            if i != j and matriks[i][j] != 0:
                return False
    return True

def cek_matriks_diagonal_rekursif(matriks, i=0, j=0):
    if i == len(matriks):
        return True

    if j == len(matriks):
        return cek_matriks_diagonal_rekursif(matriks, i + 1, 0)

    if i != j and matriks[i][j] != 0:
        return False

    return cek_matriks_diagonal_rekursif(matriks, i, j + 1)

def buat_matriks_diagonal_acak(ukuran, min_val=1, max_val=2):
    matriks = [[0 for _ in range(ukuran)] for _ in range(ukuran)]
    for i in range(ukuran):
        matriks[i][i] = random.randint(min_val, max_val)
    return matriks

def grafik():
    ukuran_matriks = range(3, 110)
    waktu_iteratif = []
    waktu_rekursif = []

    for ukuran in ukuran_matriks:
        matriks = buat_matriks_diagonal_acak(ukuran)

        mulai = time.time()
        cek_matriks_diagonal_iteratif(matriks)
        waktu_iteratif.append((time.time() - mulai) * 1e6)

        mulai = time.time()
        cek_matriks_diagonal_rekursif(matriks)
        waktu_rekursif.append((time.time() - mulai) * 1e6)

    plt.figure(figsize=(10, 6))
    plt.plot(ukuran_matriks, waktu_iteratif, label='Iteratif', marker='v')
    plt.plot(ukuran_matriks, waktu_rekursif, label='Rekursif', marker='s')
    plt.title('Worst Case Cek Matriks Diagonal')
    plt.xlabel('Ukuran Matriks (n x n)')
    plt.ylabel('Waktu (Mikrodetik)')
    plt.legend()
    plt.grid()
    plt.show()

grafik()
```

###Iterative Graph

```python
import random
import time
import matplotlib.pyplot as plt

def cek_matriks_diagonal_iteratif(matriks):
    n = len(matriks)
    for i in range(n):
        for j in range(n):
            if i != j and matriks[i][j] != 0:
                return False
    return True

def buat_matriks_diagonal_acak(ukuran, min_val=1, max_val=2):
    matriks = [[0 for _ in range(ukuran)] for _ in range(ukuran)]
    for i in range(ukuran):
        matriks[i][i] = random.randint(min_val, max_val)
    return matriks

def buat_matriks(ukuran, min_val=-10, max_val=10):
    return [[random.randint(min_val, max_val) for _ in range(ukuran)] for _ in range(ukuran)]

def buat_matriks_nondiagonal(ukuran, min_val=1, max_val=10):
    return [[random.randint(min_val, max_val) for _ in range(ukuran)] for _ in range(ukuran)]

def grafik_iteratif():
    ukuran_matriks = range(3, 110)
    waktu_iteratif_best = []
    waktu_iteratif_avg = []
    waktu_iteratif_worst = []

    for ukuran in ukuran_matriks:
        # Worst case
        matriks_worst = buat_matriks_diagonal_acak(ukuran)
        mulai = time.time()
        cek_matriks_diagonal_iteratif(matriks_worst)
        waktu_iteratif_worst.append((time.time() - mulai) * 1e6)

        # Average case
        total_iteratif = 0
        for _ in range(40):
            matriks_avg = buat_matriks(ukuran)
            mulai = time.time()
            cek_matriks_diagonal_iteratif(matriks_avg)
            total_iteratif += (time.time() - mulai) * 1e6
        waktu_iteratif_avg.append(total_iteratif / 40)

        # Best case
        matriks_best = buat_matriks_nondiagonal(ukuran)
        mulai = time.time()
        cek_matriks_diagonal_iteratif(matriks_best)
        waktu_iteratif_best.append((time.time() - mulai) * 1e6)

    plt.figure(figsize=(10, 6))
    plt.plot(ukuran_matriks, waktu_iteratif_best, label='Iteratif - Best Case', marker='v')
    plt.plot(ukuran_matriks, waktu_iteratif_avg, label='Iteratif - Average Case', marker='s')
    plt.plot(ukuran_matriks, waktu_iteratif_worst, label='Iteratif - Worst Case', marker='x')
    plt.title('Iteratif: Cek Matriks Diagonal - Best, Average, and Worst Cases')
    plt.xlabel('Ukuran Matriks (n x n)')
    plt.ylabel('Waktu (Mikrodetik)')
    plt.legend()
    plt.grid()
    plt.show()

grafik_iteratif()
```
